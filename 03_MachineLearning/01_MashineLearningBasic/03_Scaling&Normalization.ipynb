# Feature Scaling & Normalization
- 주로 수치 데이터를 대상으로 하는 작업
- 데이터의 분포 격차가 클 경우, 다른 데이터에 비해 불필요하게 강조될 수 있으므로, 불필요하게 강조될 데이터를 보정해주는 것

## ready Target

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
```

## StandardScaler

### Instance

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

## Before Scaling

```python
print(iris_df.head())
print('feature 들의 평균 값')
print(iris_df.mean())
print('\nfeature들의 분산 값')
print(iris_df.var())
```

### Scaling

```python
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df) # returnType numpy ndarray
iris_scaled
```

### NDarray To DataFrame

```python
iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
print(iris_df_scaled)
```

## MinMaxScaler

### Instance

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
iris_df_scaled = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
print(iris_df_scaled)
print('feature들의 최소값')
print(iris_df_scaled.min())
print('\nfeature들의 최대값')
print(iris_df_scaled.max())
```
