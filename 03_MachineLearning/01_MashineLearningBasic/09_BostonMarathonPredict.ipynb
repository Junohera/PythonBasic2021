```python
marathoon_2015 = pd.read_csv('../../02_DataAnalysis/marathon/marathon_results_2015.csv')
marathoon_2016 = pd.read_csv('../../02_DataAnalysis/marathon/marathon_results_2016.csv')
marathoon_2017 = pd.read_csv('../../02_DataAnalysis/marathon/marathon_results_2017.csv')


marathoon_2015['Year'] = '2015'
marathoon_2016['Year'] = '2016'
marathoon_2017['Year'] = '2017'

marathon_all = pd.concat(
    [
        marathoon_2015,
        marathoon_2016,
        marathoon_2017,
    ], ignore_index=True
)
```
