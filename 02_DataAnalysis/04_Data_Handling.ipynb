### Pandas 시작 - 파일을 DataFrame으로 불러오기, 기본 API
- read_csv()
    - read_csv()를 이용하여 csv파일을 편리하게 DataFrame으로 로딩
    - read_csv()의 sep인자를 ","가 아닌 다른 분리자로 변경하여 다른 유형의 파일도 불러오기가 가능

```python
import pandas as pd
titanicDF = pd.read_csv('titanic_train.csv')
print("type : ", type(titanicDF))
```

### head(), tail()

- DataFrame의 맨 앞 일부 데이터만 추출
- 괄호안에 원하는 수를 지정(default: 5)
- .tail()은 맨끝 일부 데이터
- tail도 head도 안쓰고, 데이터프레임 변수만 쓰고 실행하면, 모든 데이터가 추출, 많을경우 중간은 (...)로 표시

### print사용, 미사용
- print X
    - 장점 : 대상 데이터의 각 데이터를 구분하기 좋은 디자인으로 출력
    - 단점 : 한개의 셀에 하나의 출력(마지막출력)만 표시됩니다
- print O
    - 장점 : 쓰여진 만큼 모든 출력을 표시
    - 단점 : 보기 어렵게 나옴.
- print X, O 보완이 head와 tail
    - head()
    - tail()

```python
# print명령없이 변수 또는 데이터프레임의 이름만 쓰면 아래와 같이 출력
titanicDF
# a = 10
# a
```

## DataFrame에서 Series 추출 및 DataFrame 필터링 추출
- 필드명을 기준으로 다수의 값들로 이루어진 한 열의 데이터를 Series라고 부르며
- 이 Series들이 모여 DataFrame을 구성한다

```python
# DataFrame객체에서 []연산자내에 한개의 컬럼만 입력하면 Series객체를 반환
s = titanicDF['Name']
print(s.head(5))
print("## type: ", type(s))
```

```python
# DataFrame객체에서 [] 연산자내에 여러개의 컬럼명을 리스트로 입력하면
# 그 컬럼들로 구성된 DataFrame 반환
filtered_df = titanicDF[['Name', 'Age']]
print(filtered_df.head(5))
print("## type: ", type(filtered_df))
```

### describe()
- 데이터값들의 평균, 표준편차, 4분위 분포도를 제공합니다. 숫자형 컬럼들에 대해서 해당 정보 등을 제공합니다.
- 숫자데이터만

* 'PassengerId' : 탑승자 데이터 일련번호
* 'Survived' : 생존 여부(0=사망, 1=생존)
* 'Pclass' : 티켓의 선실 등급(1=일등석, 2=이등석, 3=삼등석)
* 'Name' : 탑승자 이름
* 'Sex' : 탑승자 성별
* 'Age' : 탑승자 나이
* 'SibSp' : 같이 탑승한 형제자매 또는 배우자 인원수
* 'Parch' : 같이 탑승한 부모님 또는 어린이 인원수
* 'Ticket' : 티켓 번호
* 'Fare' : 요금
* 'Cabin' : 선실 번호
* 'Embarked' : 중간 정착 항구 (C=Cherbourg, Q=Queenstown, S=Southampton)

```python
titanicDF.describe()
```

### value_count()
- 동일한 개별 데이터 값이 몇건이 있는지 정보를 제공
- 개별 데이터값의 분포도
- 주의할 점은 value_counts()는 Series객체에서만 호출될 수 있으므로 반드시 DataFrame을 단일 컬럼으로 입력하여 Series로 변환한 뒤 호출

```python
titanicDF.value_counts()
```
