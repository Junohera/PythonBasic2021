### groupby()
- groupby 인자에 Group By 하고자 하는 컬럼을 입력.
- 여러개의 컬럼으로 group By 하고자 하면 [] 내에 해당 컬럼명을 입력.
- DataFrame에 groupby()를 호출하면 DataFrameGroupBy 객체를 **리턴**합니다.

```python
# by에 입력된 'Pclass' 의 구성값들로 그룹이 만들어집니다.
titanic_groupby = titanicDF.groupby(by='Pclass')
# 현재는 그룹만 구성하고 아무일도 안한 상태이므로 
# titanic_groupby를 출력해도 결과는 Object형태로 출력됩니다.

print(titanic_groupby)
print(type(titanic_groupby))
```

```python
# 객체상태에서는 아무정보를 얻어낼 수 없습니다.
# 그룹객체를 대상으로한 함수 실행등으로 필요한 정보를 얻어냅니다.
# DataFrameGroupBy객체에 함수를 호출하여 Group By를 수행합니다.
titanic_groupby = titanicDF.groupby('Pclass').mean()
titanic_groupby
```

```python
titanic_groupby = titanicDF.groupby('Pclass').count()
titanic_groupby
```

```python
print(titanic_groupby.shape)
print(titanic_groupby.index)
```

```python
# 선실 등급별 age의 평균
titanicDF.groupby(by='Pclass')['Age', 'Survived'].mean()
```

```python
# 선실등급별 생존자의 인원수를 보고자 한다면
titanicDF.groupby(by='Pclass')[['Survived']].sum()
```

```python
# 람다를 이용한 함수정의와 사용 1 (대입 후 호출)
lambda_square = lambda x : x**2
print('3의 제곱은 : ', lambda_square(3))
```

### 연습문제
- 'Age_4Level'이란 컬럼에 Age별로 저장하고, 레벨별 인원, 생존자, 생존율을 화면에 출력하시오
    - 0~12: 어린이 (Child)
    - 13~18: 청소년 (Youth)
    - 19~55: 성년(Adult)
    - 56~: 노년 (Elderly)

```python
# 1. 데이터 불러오기
# import pandas as pd
# titanicDF = pd.read_csv('titanic_train.csv')


# 2. 나이별 레벨 부여
titanicDF['Age_4Level'] = titanicDF.Age.apply(lambda x:
    'Child' if x < 13 else (
    'Youth' if x < 19 else (
    'Adult' if x < 56 else 'Elderly'))
)

# 3. 레벨별로 그루핑
groupBy_Age_4Level = titanicDF.groupby(by='Age_4Level') # groupBy_Age_4Level에 Age_4Level로 groupby한 데이터를 저장

# 4. 그루핑된 값 중 필요값 추출
output = groupBy_Age_4Level.Survived.agg(['count', 'sum', 'mean']) # groupBy_Age_4Level에 총인원, 생존자, 생존율을 출력

# 5. 출력 포맷맞추기
output.columns = ['탑승자 인원', '생존 인원', '생존율(%)']

output['생존율(%)'] = (output['생존율(%)'] * 1000).round() / 10

output
```
