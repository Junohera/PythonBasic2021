# 데이터 결측치
1. #### isnull과 sum을 이용하여 컬럼별 결측치를 표하합니다
1. #### 각 컬럼의 이름을 출력합니다.
1. #### drop을 이용하여 'Unnamed: 0', 'Bib', 'Unnamed: 9' 컬럼 삭제
1. #### 'Senior'라는 컬럼을 추가하여 Age가 60이상인 곳에 True, 아닌곳에 False 저장

```python
# [1]
maraData.isna().count() # -> 전부 채워져있음 --> 결측치 X
```

```python
# [2]
for col in maraData.columns :
    print(col)
```

```python
# [3]
delColumns = ['Unnamed: 0', 'Bib', 'Unnamed: 9']
maraData.drop(columns=delColumns, inplace=True)
```

```python
# [4]
maraData['Senior'] = maraData.Age.apply(lambda age: age >= 60)
# groupbySenior = maraData.groupby(by='Senior')
# groupbySenior.count()
```

```python
# [5]
maraData['Year'] = '2017'
```
