```python
import pandas as pd
import numpy as np

pd.options.display.max_columns=25

marathon_2015 = pd.read_csv('./marathon/marathon_results_2015.csv')
marathon_2016 = pd.read_csv('./marathon/marathon_results_2016.csv')
marathon_2017 = pd.read_csv('./marathon/marathon_results_2017.csv')
```

```python
# 참가 년도 컬럼 추가
marathon_2015['Year'] = '2015'
marathon_2016['Year'] = '2016'
marathon_2017['Year'] = '2017'
```

## **Merge Data**

```python
marathon_merge = pd.concat([marathon_2015, marathon_2016, marathon_2017], ignore_index=True)
marathon_merge.head()
```

## **Check Data**

```python
marathon_merge.info()
```

## **Check Missing Data**

```python
marathon_merge.isnull().count()
```

```python
delColumns = ['Unnamed: 0', 'Bib', 'Unnamed: 9', 'Citizen', 'Proj Time', 'Unnamed: 8']
marathon_merge = marathon_merge.drop(columns=delColumns, axis=1)
```

```python
marathon_merge.head()
```

```python
# str 00:00:00 -> timedelta -> float64 -> int64
marathon_merge['5K'] = pd.to_timedelta(marathon_merge['5K']).astype('m8[s]').astype(np.int64)
marathon_merge['10K'] = pd.to_timedelta(marathon_merge['10K']).astype('m8[s]').astype(np.int64)
marathon_merge['15K'] = pd.to_timedelta(marathon_merge['15K']).astype('m8[s]').astype(np.int64)
marathon_merge['20K'] = pd.to_timedelta(marathon_merge['20K']).astype('m8[s]').astype(np.int64)
marathon_merge['Half'] = pd.to_timedelta(marathon_merge['Half']).astype('m8[s]').astype(np.int64)
marathon_merge['25K'] = pd.to_timedelta(marathon_merge['25K']).astype('m8[s]').astype(np.int64)
marathon_merge['30K'] = pd.to_timedelta(marathon_merge['30K']).astype('m8[s]').astype(np.int64)
marathon_merge['35K'] = pd.to_timedelta(marathon_merge['35K']).astype('m8[s]').astype(np.int64)
marathon_merge['40K'] = pd.to_timedelta(marathon_merge['40K']).astype('m8[s]').astype(np.int64)
marathon_merge['Pace'] = pd.to_timedelta(marathon_merge['Pace']).astype('m8[s]').astype(np.int64)
marathon_merge['Official Time'] = pd.to_timedelta(marathon_merge['Official Time']).astype('m8[s]').astype(np.int64)
```
