```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = 25
plt.rcParams['figure.figsize'] = [40, 10] # 전역으로 차트의 크기 설정
def setTick():
    plt.tick_params(colors='white', which='both', labelsize=40)

# 한글 폰트 설정 & 폰트가 선명하게 보이도록 retina설정
plt.rc('font', family="Malgun Gothic")
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")

fontdictHead={'fontsize': 50, 'color': 'yellow'}
fontdictBody={'fontsize': 300, 'color': 'white'}
```

```python
marathonDf = pd.read_csv('./marathon/marathon_merge[2015~2017].csv')
marathonDf.head()
```

### 남여 참가자의 연령별 기록 비교

```python
# 남자 참가자와 여자 참가자를 두개의 데이터프레임으로 따로 저장
maleRunner = marathonDf.loc[marathonDf['M/F'] == 'M', :]
femaleRunner = marathonDf.loc[marathonDf['M/F'] == 'F', :]

# 남자 참가자의 기록
y_male = maleRunner['Official Time']
# 여자 참가자의 기록
y_female = femaleRunner['Official Time']

# 남자 참가자의 나이
x_male = maleRunner['Age']

# 여자 참가자의 나이
x_female = femaleRunner['Age']

plt.plot(x_male, y_male, 's', color='b', alpha=0.3)
plt.plot(x_female, y_female, 's', color='r', alpha=0.3)
plt.title("Distribution by Running Time and Age", fontdict=fontdictHead)
plt.xlabel("Age", fontdict=fontdictHead)
plt.ylabel("Official Time", fontdict=fontdictHead)
plt.tick_params(colors='white', which='both', labelsize=40)
plt.show()
```
