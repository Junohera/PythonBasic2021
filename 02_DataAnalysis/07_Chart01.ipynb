## **Bar Chart**
- ###### 파이썬 외부 라이브러리 중 차트를 지원하는 모듈 : matplotlib.pyplot
- ###### 파이썬 차트에 얹혀서 사용되는 업그레이드 차트 모듈 : seaborn

```python
# 미국 참가자의 연령별 참석 인원수
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('dark_background') # theme

# 차트의 크기를 설정 (크기보단 비율로 사용하는 경우가 많음.)
plt.figure(figsize=(30, 10))
# countplot : seaborn의 BarChart를 생성하는 함수이며, 주로 대상 컬럼의 값 갯수를 대상으로 차트를 그립니다.
# target data - runner_USA
# count Target : 'Age'
runner_Age = sns.countplot('Age', data=runner_USA) # 차트의 큰트롤변수

# 제목
runner_Age.set_title("Number of Runner by Age", fontsize=20)
# x축
runner_Age.set_xlabel("Age", fontdict={'size': 16})
# y축
runner_Age
```

* runner_USA 데이터에 Age_10 컬럼을 추가하고
* 15세는 10대, 23세는 20대, 38세는 30대 ... 으로 입력한 후
* 연령대별 참석자 인원을 차트로 표현하세요

```python
runner_USA['Age_10'] = (runner_USA['Age'] // 10) * 10

plt.figure(figsize=(30, 10))
runner_Age_10 = sns.countplot('Age_10', data=runner_USA)

# 제목
runner_Age_10.set_title("Number of Runner by Age 10", fontdict={'size': 16, 'color': 'white'})
# x축
runner_Age_10.set_xlabel("Age grade", fontdict={'size': 16, 'color': 'white'})
# y축
runner_Age_10.set_ylabel('Number of Runner', fontdict={'size': 16, 'color': 'white'})

plt.show()
```

```python
plt.figure(figsize=(30, 10))
runner_Age_10 = sns.countplot('Age_10', data=runner_USA, hue='M/F', palette={'M':'b', 'F':'r'})

runner_Age_10.set_title("Number of Runner by Age 10, Gender - USA", fontdict={'size': 16, 'color': 'white'})
runner_Age_10.set_xlabel("Age grade", fontdict={'size': 16, 'color': 'white'})
runner_Age_10.set_ylabel('Number of Runner', fontdict={'size': 16, 'color': 'white'})

plt.show()
```

```python
# 연도별 - 연령대별 참가자 인원수 차트 표시

runner_by_year_age = marathonDf
runner_by_year_age['Age_10'] = (runner_by_year_age['Age'] // 10) * 10

# runner_by_year_age_groupby = runner_by_year_age.groupby(by=['Year', 'Age_10'])

# runner_by_year_age.info()
```

```python
plt.figure(figsize=(30, 10))
runner_by_year_age_groupby_graph = sns.countplot('Age_10', data=runner_by_year_age, hue='Year', palette={2015:'w', 2016:'b', 2017:'m'})

# 제목
runner_by_year_age_groupby_graph.set_title("RUNNER BY YEAR, AGE, Gender - USA", fontdict={'size': 16})
runner_by_year_age_groupby_graph.set_xlabel("AGE", fontdict={'size': 16})
runner_by_year_age_groupby_graph.set_ylabel('Number of Runner', fontdict={'size': 16})

plt.show()
```
