```python
from tkinter import *
from tkinter import ttk
```

```python
def btnClick1():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "1")
    
def btnClick2():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "2")
    
def btnClick3():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "3")
    
def btnClick4():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "4")
    
def btnClick5():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "5")
    
def btnClick6():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "6")
    
def btnClick7():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "7")
    
def btnClick8():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "8")
    
def btnClick9():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "9")
    
def btnClick0():
    if str1.get() == "0":
        textbox.delete(0, "")
    textbox.insert("", "0")
    
    
def btnClickC():
    global firstNum
    global op
    op = ""
    firstNum = ""
    textbox.delete(0, "")
    
def btnClickresult():
    print(firstNum + "  " + op + " " + str1.get())
    
    if firstNum == "" or str1.get() == "" :
        print("입력 이 맞지않음")
    
    left = int(firstNum)
    right = int(str1.get())
    
    result = 0
    if op == "+":
        result = left + right
    elif op == "-":
        result = left - right
    elif op == "x":
        result = left * right
    elif op == "/":
        result = left / right
        
    textbox.delete(0, "")
    textbox.insert("", str(result))
    
    
def btnClickPlus(text):
    global firstNum
    global op
    firstNum = text
    op = "+"
    textbox.delete(0, "")
    textbox.insert("", "0")
    
def btnClickMinus(text):
    global firstNum
    global op
    firstNum = text
    op = "-"
    textbox.delete(0, "")
    textbox.insert("", "0")
    
def btnClickx(text):
    global firstNum
    global op
    firstNum = text
    op = "x"
    textbox.delete(0, "")
    textbox.insert("", "0")

def btnClickslash(text):
    global firstNum
    global op
    firstNum = text
    op = "/"
    textbox.delete(0, "")
    textbox.insert("", "0")
```

```python
main = Tk()
main.title("계산기")
main.geometry()
label = Label(main, text="버튼을 이용한 계산기")
label.config(font=('굴림', 18))
label.grid(row= 0, column = 0, columnspan=4)

str1 = StringVar()
textbox = ttk.Entry(main, width=40, textvariable=str1)
textbox.config(font=('굴림', 18))
textbox.config(justify='right')
textbox.grid(row=1, column=0, columnspan=4)
textbox.insert("", "0")

btn1 = Button(main, text='1', width=10, command=lambda:btnClick1())
btn1.config(font=('굴림', 18))
btn1.grid(row=2, column=0, columnspan=1)
btn2 = Button(main, text='2', width=10, command=lambda:btnClick2())
btn2.config(font=('굴림', 18))
btn2.grid(row=2, column=1, columnspan=1)
btn3 = Button(main, text='3', width=10, command=lambda:btnClick3())
btn3.config(font=('굴림', 18))
btn3.grid(row=2, column=2, columnspan=1)
btnPlus = Button(main, text='+', width=10, command=lambda:btnClickPlus(str1.get()))
btnPlus.config(font=('굴림', 18))
btnPlus.grid(row=2, column=3, columnspan=1)
btn4 = Button(main, text='4', width=10, command=lambda:btnClick4())
btn4.config(font=('굴림', 18))
btn4.grid(row=3, column=0, columnspan=1)
btn5 = Button(main, text='5', width=10, command=lambda:btnClick5())
btn5.config(font=('굴림', 18))
btn5.grid(row=3, column=1, columnspan=1)
btn6 = Button(main, text='6', width=10, command=lambda:btnClick6())
btn6.config(font=('굴림', 18))
btn6.grid(row=3, column=2, columnspan=1)
btnMinus = Button(main, text='-', width=10, command=lambda:btnClickMinus(str1.get()))
btnMinus.config(font=('굴림', 18))
btnMinus.grid(row=3, column=3, columnspan=1)
btn7 = Button(main, text='7', width=10, command=lambda:btnClick7())
btn7.config(font=('굴림', 18))
btn7.grid(row=4, column=0, columnspan=1)
btn8 = Button(main, text='8', width=10, command=lambda:btnClick8())
btn8.config(font=('굴림', 18))
btn8.grid(row=4, column=1, columnspan=1)
btn9 = Button(main, text='9', width=10, command=lambda:btnClick9())
btn9.config(font=('굴림', 18))
btn9.grid(row=4, column=2, columnspan=1)
btnx = Button(main, text='x', width=10, command=lambda:btnClickx(str1.get()))
btnx.config(font=('굴림', 18))
btnx.grid(row=4, column=3, columnspan=1)
btnC = Button(main, text='C', width=10, command=lambda:btnClickC())
btnC.config(font=('굴림', 18))
btnC.grid(row=5, column=0, columnspan=1)
btn0 = Button(main, text='0', width=10, command=lambda:btnClick0())
btn0.config(font=('굴림', 18))
btn0.grid(row=5, column=1, columnspan=1)
btnresult = Button(main, text='=', width=10, command=lambda:btnClickresult())
btnresult.config(font=('굴림', 18))
btnresult.grid(row=5, column=2, columnspan=1)
btnslash = Button(main, text='/', width=10, command=lambda:btnClickslash(str1.get()))
btnslash.config(font=('굴림', 18))
btnslash.grid(row=5, column=3, columnspan=1)

main.mainloop()
```
