```python
import face_recognition
```

```python
import cv2
import pickle
```

```python
# 학습 데이터 경로 및 파일 이름
dataset_paths = ['dataset/son/', 'dataset/tedy/']
names = ['Son', 'Tedy']
```

```python
#  이미지 갯수와 파일 형식 저장
number_images = 10
image_type = '.jpg'
```

```python
# 학습 데이터 저장 파일명, 적용할 딥러닝 모델 방식 설정
encoding_file = 'encodings.pickle'
model_method = 'cnn'
```

```python
knownEncodings = []
knownNames = []
# 이미지 경로 갯수 만큼 반복실행
for (i, dataset_path) in enumerate(dataset_paths):
    # 현재경로에 맞는 이름 추출
    name = names[i]
    # 경로 상에서 10번(저정해놓은 파일의 갯수) 반복실행
    for idx in range(number_images):
        # 반복실행 인수로 파일경로와 이름 조립
        # "dataset/son/" + "1" + ".jpg"
        file_name = dataset_path + str(idx+1) + image_type
        
        # 생성된 경로와 파일 이름으로 파일을 읽어옵니다
        image = cv2.imread(file_name)
        
        # 이미지 색상 모드 변경
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # 얼굴 인식 실행(1. 얼굴 위치정보들 추출)
        boxes = face_recognition.face_locations(rgb, model=model_method)
        
        # 인식된 얼굴들을 인코딩 및 저장(encoding) -  인코딩된 얼굴정보가 리스트형태로
        # 저장됩니다
        encodings = face_recognition.face_encodings(rgb, boxes)
        
        # 인코딩 정보들을 다시 꺼내어 이름과 함꼐 따로 저장합니다
        for encoding in encodings:
            print(file_name, name, encoding)
            knownEncodings.append(encoding)
            knownNames.append(name)
            
data = {"encodings": knownEncodings, "names": knownNames}
f = open(encoding_file, "wb")
f.write(pickle.dumps(data))
f.close()
```
